```python
import numpy as np
```

```python
def random_list(size):
  list = []
  for i in range(size):
    list.append(np.random.randint(10,50))
  return list

def sum_random (list):
  sum = 0
  for i in range(len(list)):
    sum = sum + list[i]
  return sum
```

```python
n = int(input("Enter some integer between 5 and 15 "))
if n < 5 or n > 15:
  print("Please make sure the integer entered is between 5 and 15")
```

```python
MyList = random_list(n)
MyList
```

```python
MySum = sum_random(MyList)
MySum
```
